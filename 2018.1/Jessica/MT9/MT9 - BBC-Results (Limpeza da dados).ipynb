```python
import pandas as pd

dataset = pd.read_csv('BBC-Results.csv',sep=',')
dataset.columns = ['Timestamp', 'How many hours did you sleep last night?', 'Recognition Score', 'Temporal Memory Score', 'coluna4', 'coluna5']
dataset
```

```python
dataset.isnull().sum()
```

```python
dataset = dataset.drop('coluna4', axis=1)
dataset = dataset.drop('coluna5', axis=1)
dataset
```

```python
#Convertendo valores em strings para inteiro (Dessa forma fica mais fácil preencher os campos nulos com a média da coluna)
dataset['How many hours did you sleep last night?'] = pd.to_numeric(dataset['How many hours did you sleep last night?'], errors='coerce')
dataset['Recognition Score'] = pd.to_numeric(dataset['Recognition Score'], errors='coerce')
dataset['Temporal Memory Score'] = pd.to_numeric(dataset['Temporal Memory Score'], errors='coerce')
dataset
```

```python
#Preenchendo os campos nulos com a média da coluna pertecente
dataset['How many hours did you sleep last night?'].fillna(dataset['How many hours did you sleep last night?'].mean(), inplace=True)
dataset['Recognition Score'].fillna(dataset['Recognition Score'].mean(), inplace=True)
dataset['Temporal Memory Score'].fillna(dataset['Temporal Memory Score'].mean(), inplace=True)
dataset
```

```python
dataset.isnull().sum()
```

```python
len( dataset[ dataset['How many hours did you sleep last night?'] > 24])
```

```python
len( dataset[ dataset['Recognition Score'] > 100])
```

```python
len( dataset[ dataset['Temporal Memory Score'] > 100])
```

```python
dataset.to_csv('BBC-Results.csv', sep=',', index= False)
```
