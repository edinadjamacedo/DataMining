## 1. Analisando os relatórios de todos os anos. Qual é a área que arca mais custos e qual apresenta o maior indice crescente?

A área que tem os maiores gastos em todos os anos é a de "Bolsas e Auxilios", onde o custo é aproximadamente 1/3 do total, assim como também é a que apresenta o maior indice crescente. Como você pode ver no gráfico abaixo.

```python
import matplotlib.pyplot as plt

x = ['2015','2016','2017']
y = [611270, 885270, 957630]

plt.bar(x, y, label = 'Barsl', color = 'k')
plt.title('Despesas totais de "Bolsas e Auxilios"')
plt.xlabel('Anos')
plt.ylabel('Custos')


plt.show()
```
